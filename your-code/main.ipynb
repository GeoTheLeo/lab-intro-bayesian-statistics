## 1. Cookie Problem

Suppose we have two bowls of cookies. Bowl 1 contains 30 vanilla cookies and 10 chocolate cookies. Bowl 2 contains 20 of each. You randomly pick one cookie out of one of the bowls, and it is vanilla. Use Bayes Theorem to calculate the probability that the vanilla cookie you picked came from Bowl 1?

```python
# Given data
# Bowl 1: 30 vanilla, 10 chocolate → total 40 cookies
# Bowl 2: 20 vanilla, 20 chocolate → total 40 cookies

# Prior probabilities (equal chance of picking either bowl)
P_B1 = 0.5  # P(Bowl 1)
P_B2 = 0.5  # P(Bowl 2)

# Likelihoods: P(Vanilla | Bowl)
P_V_given_B1 = 30 / 40  # 0.75
P_V_given_B2 = 20 / 40  # 0.50

# Evidence: P(Vanilla) = P(V|B1)*P(B1) + P(V|B2)*P(B2)
P_V = (P_V_given_B1 * P_B1) + (P_V_given_B2 * P_B2)

# Posterior probability: P(Bowl 1 | Vanilla) using Bayes Theorem
P_B1_given_V = (P_V_given_B1 * P_B1) / P_V

print("Bayesian Cookie Bowl Analysis")
print("=" * 40)
print(f"Prior P(Bowl 1): {P_B1}")
print(f"Prior P(Bowl 2): {P_B2}")
print(f"P(Vanilla | Bowl 1): {P_V_given_B1}")
print(f"P(Vanilla | Bowl 2): {P_V_given_B2}")
print(f"P(Vanilla): {P_V}")
print(f"\nP(Bowl 1 | Vanilla) = ({P_V_given_B1} × {P_B1}) / {P_V}")
print(f"P(Bowl 1 | Vanilla) = {P_B1_given_V:.3f}")
print(f"\nAnswer: The probability the vanilla cookie came from Bowl 1 is {P_B1_given_V:.1%}")
```

## 2. Candy Problem

Suppose you have two bags of candies:

- In Bag 1, the mix of colors is:
    - Brown - 30%
    - Yellow - 20%
    - Red - 20%
    - Green - 10%
    - Orange - 10%
    - Tan - 10%
    
- In Bag 2, the mix of colors is:
    - Blue - 24%
    - Green - 20%
    - Orange - 16%
    - Yellow - 14%
    - Red - 13%
    - Brown - 13%
    
Not knowing which bag is which, you randomly draw one candy from each bag. One is yellow and one is green. What is the probability that the yellow one came from the Bag 1?

*Hint: For the likelihoods, you will need to multiply the probabilities of drawing yellow from one bag and green from the other bag and vice versa.*

```python
# Probabilities
P_Y1 = 0.5  # prior that yellow from Bag1
P_Y2 = 0.5  # prior that yellow from Bag2

# Likelihoods
P_yellow_given_B1 = 0.20
P_green_given_B2 = 0.20
P_data_given_Y1 = P_yellow_given_B1 * P_green_given_B2  # 0.04

P_yellow_given_B2 = 0.14
P_green_given_B1 = 0.10
P_data_given_Y2 = P_yellow_given_B2 * P_green_given_B1  # 0.014

# Evidence
P_data = (P_data_given_Y1 * P_Y1) + (P_data_given_Y2 * P_Y2)  # 0.02 + 0.007 = 0.027

# Posterior
P_Y1_given_data = (P_data_given_Y1 * P_Y1) / P_data

print(f"P(Yellow from Bag1 | one yellow and one green) = {P_Y1_given_data:.4f} or {P_Y1_given_data:.2%}")
```

What is the probability that the yellow candy came from Bag 2?

```python
# Given probabilities
# Bag 1: Brown 30%, Yellow 20%, Red 20%, Green 10%, Orange 10%, Tan 10%
# Bag 2: Blue 24%, Green 20%, Orange 16%, Yellow 14%, Red 13%, Brown 13%

# Prior probabilities (equal chance that yellow came from either bag)
P_Y1 = 0.5  # Prior: yellow from Bag 1
P_Y2 = 0.5  # Prior: yellow from Bag 2

# Likelihoods
# Case 1: Yellow from Bag 1, Green from Bag 2
P_yellow_B1 = 0.20
P_green_B2 = 0.20
P_data_given_Y1 = P_yellow_B1 * P_green_B2  # 0.20 * 0.20 = 0.04

# Case 2: Yellow from Bag 2, Green from Bag 1
P_yellow_B2 = 0.14
P_green_B1 = 0.10
P_data_given_Y2 = P_yellow_B2 * P_green_B1  # 0.14 * 0.10 = 0.014

# Evidence (total probability of observing one yellow and one green)
P_data = (P_data_given_Y1 * P_Y1) + (P_data_given_Y2 * P_Y2)  # (0.04*0.5) + (0.014*0.5) = 0.027

# Posterior probability that yellow came from Bag 2
P_Y2_given_data = (P_data_given_Y2 * P_Y2) / P_data

print("Bayesian Candy Bag Analysis")
print("=" * 40)
print(f"Prior P(Yellow from Bag 1): {P_Y1}")
print(f"Prior P(Yellow from Bag 2): {P_Y2}")
print(f"P(Data | Yellow from Bag 1): {P_data_given_Y1}")
print(f"P(Data | Yellow from Bag 2): {P_data_given_Y2}")
print(f"P(Data): {P_data}")
print(f"\nP(Yellow from Bag 2 | Data) = ({P_data_given_Y2} × {P_Y2}) / {P_data}")
print(f"P(Yellow from Bag 2 | Data) = {P_Y2_given_data:.4f}")
print(f"\nAnswer: The probability that the yellow candy came from Bag 2 is {P_Y2_given_data:.2%}")

# Verify probabilities sum to 1
P_Y1_given_data = (P_data_given_Y1 * P_Y1) / P_data
print(f"\nProbability check:")
print(f"P(Bag 1 | Data) + P(Bag 2 | Data) = {P_Y1_given_data:.4f} + {P_Y2_given_data:.4f} = {P_Y1_given_data + P_Y2_given_data:.4f}")
```

## 3. Monty Hall Problem

Suppose you are a contestant on the popular game show *Let's Make a Deal*. The host of the show (Monty Hall) presents you with three doors - Door A, Door B, and Door C. He tells you that there is a sports car behind one of them and if you choose the correct one, you win the car!

You select Door A, but then Monty makes things a little more interesting. He opens Door B to reveal that there is no sports car behind it and asks you if you would like to stick with your choice of Door A or switch your choice to Door C. Given this new information, what are the probabilities of you winning the car if you stick with Door A versus if you switch to Door C?

```python
import numpy as np

def monty_hall_simulation(n_simulations=100000):
    """
    Simulate the Monty Hall problem to calculate winning probabilities
    when sticking with the original choice vs switching doors.
    """
    # Initialize counters
    stick_wins = 0
    switch_wins = 0
    
    for _ in range(n_simulations):
        # Randomly place the car behind one of the three doors
        car_door = np.random.randint(1, 4)
        
        # Contestant randomly chooses a door (Door A = 1, Door B = 2, Door C = 3)
        initial_choice = np.random.randint(1, 4)
        
        # Monty opens a door that:
        # 1. Is not the contestant's choice
        # 2. Does not have the car
        possible_doors_to_open = [door for door in [1, 2, 3] 
                                if door != initial_choice and door != car_door]
        
        # If there's only one possible door to open, Monty opens it
        # If there are two (when contestant initially picked the car), Monty randomly chooses one
        door_opened = np.random.choice(possible_doors_to_open)
        
        # Determine the remaining door if switching
        remaining_door = [door for door in [1, 2, 3] 
                         if door != initial_choice and door != door_opened][0]
        
        # Check if sticking wins
        if initial_choice == car_door:
            stick_wins += 1
        
        # Check if switching wins
        if remaining_door == car_door:
            switch_wins += 1
    
    # Calculate probabilities
    stick_prob = stick_wins / n_simulations
    switch_prob = switch_wins / n_simulations
    
    return stick_prob, switch_prob

# Run simulation
n_trials = 100000
stick_prob, switch_prob = monty_hall_simulation(n_trials)

# Analytical solution
print("Monty Hall Problem - Analytical Solution")
print("=" * 40)
print("Initial choice: Door A")
print("Monty opens: Door B (no car)")
print("\nProbability if you STICK with Door A:")
print("P(Win | Stick) = P(Car behind A | Monty opened B)")
print("              = 1/3 ≈ 33.33%")

print("\nProbability if you SWITCH to Door C:")
print("P(Win | Switch) = P(Car behind C | Monty opened B)")
print("               = 2/3 ≈ 66.67%")

# Simulation results
print(f"\nSimulation Results ({n_trials:,} trials):")
print("=" * 40)
print(f"Probability of winning if you STICK: {stick_prob:.4f} ({stick_prob:.2%})")
print(f"Probability of winning if you SWITCH: {switch_prob:.4f} ({switch_prob:.2%})")

# Visualize the results
import matplotlib.pyplot as plt

strategies = ['Stick with Door A', 'Switch to Door C']
probabilities = [stick_prob, switch_prob]
colors = ['lightcoral', 'lightgreen']

plt.figure(figsize=(10, 6))
bars = plt.bar(strategies, probabilities, color=colors, alpha=0.7, edgecolor='black')

# Add value labels on bars
for bar, prob in zip(bars, probabilities):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
             f'{prob:.2%}', ha='center', va='bottom', fontweight='bold')

plt.title('Monty Hall Problem: Winning Probabilities', fontsize=16, fontweight='bold')
plt.ylabel('Probability of Winning', fontsize=12)
plt.ylim(0, 0.8)
plt.grid(axis='y', alpha=0.3)

# Add theoretical lines
plt.axhline(y=1/3, color='red', linestyle='--', alpha=0.7, label='Theoretical: 1/3')
plt.axhline(y=2/3, color='green', linestyle='--', alpha=0.7, label='Theoretical: 2/3')
plt.legend()

plt.tight_layout()
plt.show()

# Explanation of why switching is better
print("\nWhy Switching is Better:")
print("=" * 40)
print("Initial probability of choosing correctly: 1/3")
print("Initial probability of choosing wrong: 2/3")
print("\nIf you initially chose WRONG (2/3 probability):")
print("- Monty reveals the only other wrong door")
print("- Switching guarantees you win the car!")
print("\nIf you initially chose RIGHT (1/3 probability):")
print("- Monty reveals one of the two wrong doors")
print("- Switching makes you lose")
print("\nOverall: P(Win by switching) = P(Initially wrong) = 2/3")
```

```python
# Why Switching Doubles Your Chances:
# Initial choice: 1/3 chance you picked correctly

# If you were wrong initially (2/3 chance):

# Monty reveals the only other wrong door

# Switching guarantees you win!

# If you were right initially (1/3 chance):

# Switching makes you lose

# Therefore, switching gives you a 2/3 chance of winning because it turns all your initial wrong guesses into wins!

# This simulation confirms the mathematical analysis - you should always switch doors in the Monty Hall problem!
```

## 1. Bayesian Analysis 

Suppose you work for a landscaping company, and they want to advertise their service online. They create an ad and sit back waiting for the money to roll in. On the first day, the ad sends 100 visitors to the site and 14 of them sign up for landscaping services. Create a generative model to come up with the posterior distribution and produce a visualization of what the posterior distribution would look like given the observed data.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy import special

# Observed data
visitors = 100
conversions = 14

# Bayesian analysis using Beta-Binomial conjugate prior
# Prior: We'll use a weak prior Beta(1, 1) which is uniform
alpha_prior = 1
beta_prior = 1

# Posterior: Beta(alpha_prior + conversions, beta_prior + visitors - conversions)
alpha_posterior = alpha_prior + conversions
beta_posterior = beta_prior + visitors - conversions

print("Bayesian Analysis of Conversion Rate")
print("=" * 40)
print(f"Observed data: {conversions} conversions out of {visitors} visitors")
print(f"Observed conversion rate: {conversions/visitors:.2%}")
print(f"\nPrior: Beta({alpha_prior}, {beta_prior})")
print(f"Posterior: Beta({alpha_posterior}, {beta_posterior})")

# Generate posterior distribution
theta = np.linspace(0, 0.3, 1000)  # Conversion rate range from 0% to 30%
posterior_pdf = stats.beta.pdf(theta, alpha_posterior, beta_posterior)

# Calculate posterior statistics
posterior_mean = alpha_posterior / (alpha_posterior + beta_posterior)
posterior_median = special.betaincinv(alpha_posterior, beta_posterior, 0.5)
posterior_mode = (alpha_posterior - 1) / (alpha_posterior + beta_posterior - 2) if alpha_posterior > 1 and beta_posterior > 1 else posterior_mean

# Calculate credible interval
credible_interval = stats.beta.interval(0.95, alpha_posterior, beta_posterior)

print(f"\nPosterior Statistics:")
print(f"Mean conversion rate: {posterior_mean:.3f} ({posterior_mean:.2%})")
print(f"Median conversion rate: {posterior_median:.3f} ({posterior_median:.2%})")
print(f"Mode conversion rate: {posterior_mode:.3f} ({posterior_mode:.2%})")
print(f"95% Credible Interval: [{credible_interval[0]:.3f}, {credible_interval[1]:.3f}]")
print(f"95% Credible Interval: [{credible_interval[0]:.2%}, {credible_interval[1]:.2%}]")

# Create the visualization
plt.figure(figsize=(12, 8))

# Plot posterior distribution
plt.plot(theta, posterior_pdf, 'b-', linewidth=2, label='Posterior Distribution')
plt.fill_between(theta, posterior_pdf, alpha=0.3, color='blue')

# Add mean, median, mode lines
plt.axvline(posterior_mean, color='red', linestyle='--', linewidth=2, label=f'Mean: {posterior_mean:.2%}')
plt.axvline(posterior_median, color='green', linestyle='--', linewidth=2, label=f'Median: {posterior_median:.2%}')
plt.axvline(posterior_mode, color='orange', linestyle='--', linewidth=2, label=f'Mode: {posterior_mode:.2%}')

# Add credible interval
plt.axvline(credible_interval[0], color='purple', linestyle=':', linewidth=2, label='95% Credible Interval')
plt.axvline(credible_interval[1], color='purple', linestyle=':', linewidth=2)

# Add observed conversion rate
observed_rate = conversions / visitors
plt.axvline(observed_rate, color='black', linestyle='-', linewidth=2, label=f'Observed: {observed_rate:.2%}')

# Customize the plot
plt.title('Posterior Distribution of Conversion Rate\n(Bayesian Analysis with Beta-Binomial)', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Conversion Rate', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)

# Add text box with statistics
stats_text = f"""Data: {conversions}/{visitors} = {observed_rate:.1%}
Posterior: Beta({alpha_posterior}, {beta_posterior})
Mean: {posterior_mean:.2%}
95% CI: [{credible_interval[0]:.2%}, {credible_interval[1]:.2%}]"""

plt.text(0.02, 0.98, stats_text, transform=plt.gca().transAxes, 
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8),
         fontsize=10)

plt.tight_layout()
plt.show()

# Additional: Prior vs Posterior comparison
plt.figure(figsize=(10, 6))

# Prior distribution
prior_pdf = stats.beta.pdf(theta, alpha_prior, beta_prior)

plt.plot(theta, prior_pdf, 'r-', linewidth=2, label='Prior Distribution (Uniform)')
plt.plot(theta, posterior_pdf, 'b-', linewidth=2, label='Posterior Distribution')

plt.fill_between(theta, prior_pdf, alpha=0.2, color='red')
plt.fill_between(theta, posterior_pdf, alpha=0.2, color='blue')

plt.title('Prior vs Posterior Distribution of Conversion Rate', fontsize=14, fontweight='bold')
plt.xlabel('Conversion Rate', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Predictive distribution for future visitors
print(f"\nPredictive Analysis:")
print("If we get 100 more visitors, we expect:")
print(f"Expected conversions: {posterior_mean * 100:.1f}")
print(f"95% prediction interval: [{credible_interval[0] * 100:.1f}, {credible_interval[1] * 100:.1f}] conversions")
```

Produce a set of descriptive statistics for the posterior distribution.

```python
import numpy as np
import scipy.stats as stats
from scipy import special
import matplotlib.pyplot as plt

# Observed data
visitors = 100
conversions = 14

# Posterior parameters (using Beta(1,1) prior)
alpha_post = 1 + conversions
beta_post = 1 + visitors - conversions

print("DESCRIPTIVE STATISTICS FOR POSTERIOR DISTRIBUTION")
print("=" * 50)
print(f"Data: {conversions} conversions out of {visitors} visitors")
print(f"Posterior: Beta({alpha_post}, {beta_post})")
print("=" * 50)

# 1. Central Tendency Measures
print("\n1. CENTRAL TENDENCY")
print("-" * 30)

# Mean
posterior_mean = alpha_post / (alpha_post + beta_post)
print(f"Mean: {posterior_mean:.4f} ({posterior_mean:.2%})")

# Median
posterior_median = special.betaincinv(alpha_post, beta_post, 0.5)
print(f"Median: {posterior_median:.4f} ({posterior_median:.2%})")

# Mode
if alpha_post > 1 and beta_post > 1:
    posterior_mode = (alpha_post - 1) / (alpha_post + beta_post - 2)
else:
    posterior_mode = posterior_mean
print(f"Mode: {posterior_mode:.4f} ({posterior_mode:.2%})")

# 2. Dispersion Measures
print("\n2. DISPERSION")
print("-" * 30)

# Variance
posterior_variance = (alpha_post * beta_post) / ((alpha_post + beta_post)**2 * (alpha_post + beta_post + 1))
print(f"Variance: {posterior_variance:.6f}")

# Standard Deviation
posterior_std = np.sqrt(posterior_variance)
print(f"Standard Deviation: {posterior_std:.4f} ({posterior_std:.2%})")

# Coefficient of Variation
cv = posterior_std / posterior_mean
print(f"Coefficient of Variation: {cv:.4f} ({cv:.2%})")

# 3. Shape Measures
print("\n3. SHAPE")
print("-" * 30)

# Skewness
skewness = (2 * (beta_post - alpha_post) * np.sqrt(alpha_post + beta_post + 1)) / \
           ((alpha_post + beta_post + 2) * np.sqrt(alpha_post * beta_post))
print(f"Skewness: {skewness:.4f}")
print("  → Right-skewed" if skewness > 0 else "  → Left-skewed" if skewness < 0 else "  → Symmetric")

# Kurtosis
kurtosis = 6 * ((alpha_post - beta_post)**2 * (alpha_post + beta_post + 1) - 
                alpha_post * beta_post * (alpha_post + beta_post + 2)) / \
           (alpha_post * beta_post * (alpha_post + beta_post + 2) * (alpha_post + beta_post + 3))
excess_kurtosis = kurtosis - 3
print(f"Excess Kurtosis: {excess_kurtosis:.4f}")
print("  → Leptokurtic (heavy-tailed)" if excess_kurtosis > 0 else 
      "  → Platykurtic (light-tailed)" if excess_kurtosis < 0 else "  → Mesokurtic (normal-like)")

# 4. Quantiles and Intervals
print("\n4. QUANTILES AND INTERVALS")
print("-" * 30)

# Credible Intervals
ci_90 = stats.beta.interval(0.90, alpha_post, beta_post)
ci_95 = stats.beta.interval(0.95, alpha_post, beta_post)
ci_99 = stats.beta.interval(0.99, alpha_post, beta_post)

print(f"90% Credible Interval: [{ci_90[0]:.4f}, {ci_90[1]:.4f}] → [{ci_90[0]:.2%}, {ci_90[1]:.2%}]")
print(f"95% Credible Interval: [{ci_95[0]:.4f}, {ci_95[1]:.4f}] → [{ci_95[0]:.2%}, {ci_95[1]:.2%}]")
print(f"99% Credible Interval: [{ci_99[0]:.4f}, {ci_99[1]:.4f}] → [{ci_99[0]:.2%}, {ci_99[1]:.2%}]")

# Quartiles
Q1 = special.betaincinv(alpha_post, beta_post, 0.25)
Q3 = special.betaincinv(alpha_post, beta_post, 0.75)
IQR = Q3 - Q1
print(f"Quartiles: Q1 = {Q1:.4f}, Q3 = {Q3:.4f}, IQR = {IQR:.4f}")

# 5. Probability Calculations
print("\n5. PROBABILITY CALCULATIONS")
print("-" * 30)

# Probability that conversion rate is above certain thresholds
p_above_10 = 1 - stats.beta.cdf(0.10, alpha_post, beta_post)
p_above_15 = 1 - stats.beta.cdf(0.15, alpha_post, beta_post)
p_above_20 = 1 - stats.beta.cdf(0.20, alpha_post, beta_post)

print(f"P(rate > 10%): {p_above_10:.4f} ({p_above_10:.2%})")
print(f"P(rate > 15%): {p_above_15:.4f} ({p_above_15:.2%})")
print(f"P(rate > 20%): {p_above_20:.4f} ({p_above_20:.2%})")

# Probability within ranges
p_10_to_20 = stats.beta.cdf(0.20, alpha_post, beta_post) - stats.beta.cdf(0.10, alpha_post, beta_post)
print(f"P(10% < rate < 20%): {p_10_to_20:.4f} ({p_10_to_20:.2%})")

# 6. Distribution Properties
print("\n6. DISTRIBUTION PROPERTIES")
print("-" * 30)

print(f"Alpha parameter: {alpha_post}")
print(f"Beta parameter: {beta_post}")
print(f"Sample size equivalent: {alpha_post + beta_post - 2}")
print(f"Effective sample size: {alpha_post + beta_post}")

# 7. Visualization of Key Statistics
theta = np.linspace(0, 0.3, 1000)
posterior_pdf = stats.beta.pdf(theta, alpha_post, beta_post)

plt.figure(figsize=(12, 8))
plt.plot(theta, posterior_pdf, 'b-', linewidth=2, label='Posterior PDF')

# Mark key statistics
plt.axvline(posterior_mean, color='red', linestyle='-', linewidth=2, label=f'Mean: {posterior_mean:.2%}')
plt.axvline(posterior_median, color='green', linestyle='--', linewidth=2, label=f'Median: {posterior_median:.2%}')
plt.axvline(posterior_mode, color='orange', linestyle=':', linewidth=2, label=f'Mode: {posterior_mode:.2%}')

# Shade credible intervals
plt.fill_between(theta, posterior_pdf, where=(theta >= ci_95[0]) & (theta <= ci_95[1]), 
                 alpha=0.3, color='blue', label='95% Credible Interval')

plt.title('Posterior Distribution with Key Statistics', fontsize=16, fontweight='bold')
plt.xlabel('Conversion Rate', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)

# Add statistics table
stats_text = f"""Key Statistics:
Mean: {posterior_mean:.2%}
Std Dev: {posterior_std:.2%}
Skewness: {skewness:.2f}
95% CI: [{ci_95[0]:.2%}, {ci_95[1]:.2%}]
P(rate > 15%): {p_above_15:.1%}"""

plt.text(0.02, 0.98, stats_text, transform=plt.gca().transAxes, 
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8),
         fontsize=10)

plt.tight_layout()
plt.show()

# 8. Summary Table
print("\n8. SUMMARY TABLE")
print("-" * 30)
summary_data = {
    'Measure': ['Mean', 'Median', 'Mode', 'Std Dev', 'Variance', 'Skewness', 'Excess Kurtosis', 
                '90% CI Lower', '90% CI Upper', '95% CI Lower', '95% CI Upper'],
    'Value': [posterior_mean, posterior_median, posterior_mode, posterior_std, posterior_variance,
              skewness, excess_kurtosis, ci_90[0], ci_90[1], ci_95[0], ci_95[1]],
    'Percentage': [f"{posterior_mean:.2%}", f"{posterior_median:.2%}", f"{posterior_mode:.2%}", 
                   f"{posterior_std:.2%}", "-", "-", "-", f"{ci_90[0]:.2%}", f"{ci_90[1]:.2%}", 
                   f"{ci_95[0]:.2%}", f"{ci_95[1]:.2%}"]
}

for i, (measure, value, pct) in enumerate(zip(summary_data['Measure'], summary_data['Value'], summary_data['Percentage'])):
    if i in [0, 1, 2, 3, 7, 8, 9, 10]:  # Show percentage for these measures
        print(f"{measure:<15} {value:.4f} ({pct})")
    else:
        print(f"{measure:<15} {value:.4f}")
```

What is the 90% credible interval range?

```python
import scipy.stats as stats

# Posterior parameters
alpha_post = 1 + 14  # prior alpha + conversions
beta_post = 1 + 100 - 14  # prior beta + failures

# Calculate 90% credible interval
ci_90 = stats.beta.interval(0.90, alpha_post, beta_post)

print("90% CREDIBLE INTERVAL FOR CONVERSION RATE")
print("=" * 40)
print(f"Posterior distribution: Beta({alpha_post}, {beta_post})")
print(f"90% Credible Interval: [{ci_90[0]:.4f}, {ci_90[1]:.4f}]")
print(f"90% Credible Interval: [{ci_90[0]:.2%}, {ci_90[1]:.2%}]")
print(f"Interval Width: {ci_90[1] - ci_90[0]:.4f} ({(ci_90[1] - ci_90[0]):.2%})")

# Interpretation
print(f"\nInterpretation:")
print(f"We are 90% confident that the true conversion rate is between {ci_90[0]:.2%} and {ci_90[1]:.2%}")
print(f"This means there's a 5% chance the rate is below {ci_90[0]:.2%}")
print(f"And a 5% chance the rate is above {ci_90[1]:.2%}")
```

What is the Maximum Likelihood Estimate?

```python
# the real data
visitors = 100
conversions = 14

# Maximum Likelihood Estimate (MLE)
mle = conversions / visitors

print("MAXIMUM LIKELIHOOD ESTIMATE (MLE)")
print("=" * 40)
print(f"Observed data: {conversions} conversions out of {visitors} visitors")
print(f"MLE = conversions / visitors = {conversions} / {visitors}")
print(f"MLE = {mle:.4f} ({mle:.2%})")

# Compare with Bayesian posterior mean
alpha_post = 1 + conversions
beta_post = 1 + visitors - conversions
posterior_mean = alpha_post / (alpha_post + beta_post)

print(f"\nComparison with Bayesian Estimate:")
print(f"MLE (Frequentist): {mle:.4f} ({mle:.2%})")
print(f"Posterior Mean (Bayesian): {posterior_mean:.4f} ({posterior_mean:.2%})")
print(f"Difference: {abs(mle - posterior_mean):.6f} ({abs(mle - posterior_mean):.4%})")

# Why they're almost identical
print(f"\nWhy they're almost identical:")
print(f"Prior: Beta(1, 1) - this is a uniform/uninformative prior")
print(f"With large sample sizes, MLE and Bayesian estimates converge")
print(f"With only 100 visitors, there's a tiny difference due to the +1 prior")
```
